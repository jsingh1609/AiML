# Data cleaning

```python
df.isnull()  # check for null values
```

```python
print(df.isnull().sum())
df['age'].fillna(df['age'].mean(), inplace=True)  # fill nulls in 'age' with mean
```

```python
df = df.drop('deck', axis=1, errors='ignore')  # drop 'deck' column
print(df.head())
```

# FILTERING DATA

```python
adults = df[df['age'] >= 18]  # filter adults
print(adults)
female = df[df['sex']=='woman']  # filter
print(female)
```
