```python
from datetime import datetime

def longest_gap(timestamps):
    times = sorted(datetime.fromisoformat(t) for t in timestamps)
    gaps = [times[i+1] - times[i] for i in range(len(times)-1)]
    return max(gaps)

times = [
    "2024-01-01T08:00:00",
    "2024-01-01T08:05:00",
    "2024-01-01T10:00:00",
    "2024-01-01T10:03:00",
    "2024-01-01T10:04:30",
]
print(longest_gap(times))
```
