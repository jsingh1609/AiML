```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the diabetes dataset
df = pd.read_csv("diabetes.csv")

# -----------------------------
# POSITIVE LINEAR REGRESSION
# -----------------------------
# Glucose  vs  BMI  (positive correlation = +0.22)
# As Glucose goes up, BMI tends to go up too

# Dataset
X_pos = df["Glucose"].values.reshape(-1, 1)
Y_pos = df["BMI"].values

# Model
model_pos = LinearRegression()
model_pos.fit(X_pos, Y_pos)

# Prediction
Y_pred_pos = model_pos.predict(X_pos)

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(X_pos, Y_pos, color="steelblue", alpha=0.5, edgecolors="black", linewidths=0.5)
plt.plot(sorted(X_pos.flatten()),
         model_pos.predict(np.array(sorted(X_pos.flatten())).reshape(-1, 1)),
         color="red", linewidth=2)
plt.title("Positive Linear Regression    Glucose vs BMI")
plt.xlabel("Glucose")
plt.ylabel("BMI")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

print("Positive Regression Equation:")
print(f"BMI = {model_pos.coef_[0]:.4f} x Glucose + {model_pos.intercept_:.2f}")
print(f"R Score : {model_pos.score(X_pos, Y_pos):.4f}\n")

# -----------------------------
# NEGATIVE LINEAR REGRESSION
# -----------------------------
# Age  vs  SkinThickness  (negative correlation = -0.11)
# As Age goes up, SkinThickness tends to go down

# Dataset
X_neg = df["Age"].values.reshape(-1, 1)
Y_neg = df["SkinThickness"].values

# Model
model_neg = LinearRegression()
model_neg.fit(X_neg, Y_neg)

# Prediction
Y_pred_neg = model_neg.predict(X_neg)

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(X_neg, Y_neg, color="coral", alpha=0.5, edgecolors="black", linewidths=0.5)
plt.plot(sorted(X_neg.flatten()),
         model_neg.predict(np.array(sorted(X_neg.flatten())).reshape(-1, 1)),
         color="darkred", linewidth=2)
plt.title("Negative Linear Regression    Age vs SkinThickness")
plt.xlabel("Age")
plt.ylabel("SkinThickness")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

print("Negative Regression Equation:")
print(f"SkinThickness = {model_neg.coef_[0]:.4f} x Age + {model_neg.intercept_:.2f}")
print(f"R Score : {model_neg.score(X_neg, Y_neg):.4f}")
```
