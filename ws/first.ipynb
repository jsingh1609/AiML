```python
from zeep import Client, Transport
from zeep.cache import SqliteCache
from requests import Session
from zeep.wsdl.utils import etree_to_string

session = Session()
cache = SqliteCache()  # Uses default cache location (cross-platform)
transport = Transport(cache=cache, session=session)
client = Client('http://www.dneonline.com/calculator.asmx?WSDL', transport=transport)

# Build and show the raw SOAP request XML
envelope = client.create_message(client.service, "Add", intA=10, intB=20)
xml = etree_to_string(envelope)
print("SOAP Request XML:")
print(xml.decode("utf-8"))
print("\n" + "="*50 + "\n")

headers = {
    "Content-Type": "text/xml; charset=utf-8",
    "SOAPAction": "http://tempuri.org/Add",
}

response = client.transport.post_xml(
    client.service._binding_options["address"],
    envelope,
    headers
)
print("SOAP Response:")
print(response.content.decode())
print("\n" + "="*50 + "\n")

print("Result using client.service.Add():")
print(client.service.Add(10, 20))
```

```python
from zeep import Client
from zeep.wsdl.utils import etree_to_string

# 1. Setup the client (removed &quot from the URL)
client = Client("http://www.dneonline.com/calculator.asmx?WSDL")

# 2. Create the XML message object (This is an 'lxml.etree.Element')
node = client.create_message(
    client.service,
    "Add",
    intA=10,
    intB=20
)

print("Generated SOAP XML:")
print(etree_to_string(node).decode('utf-8'))
print("\n" + "="*50 + "\n")

# 3. Send the request
response = client.transport.post_xml(
    "http://www.dneonline.com/calculator.asmx",
    envelope=node,
    headers={
        "Content-Type": "text/xml; charset=utf-8",
        "SOAPAction": "http://tempuri.org/Add"
    }
)

print("Raw Response:")
print(response.content.decode())
print("\n" + "="*50 + "\n")

# 4. Or simply use the service directly
print("Direct service call result:")
result = client.service.Add(10, 20)
print(f"10 + 20 = {result}")
```
